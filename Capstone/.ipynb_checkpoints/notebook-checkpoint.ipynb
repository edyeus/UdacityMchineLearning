### Data Cleanup ###

```python
# Import libraries necessary for this project
import numpy as np
import pandas as pd
from IPython.display import display

# Pretty display for notebooks
%matplotlib inline

DATA_SET = "CTG_cleaned.xlsx"

#Importing the dataset
data = pd.read_excel(DATA_SET)
display(data.head())
print(data.columns)
```

```python
#drop 10 level classification
data_NSP_only = data

tenLevel = ['A', 'B', 'C', 'D', 'E', 'AD', 'DE', 'LD', 'FS', 'SUSP', 'CLASS']
data_NSP_only = data_NSP_only.drop(tenLevel, axis=1)
    
#drop time and filename related columns
timeCol = ['FileName', 'Date', 'SegFile', 'b', 'e']
data_NSP_only = data_NSP_only.drop(timeCol, axis=1)
    
#drop DR, since all the values are the same
data_NSP_only = data_NSP_only.drop('DR', axis=1)

print(data_NSP_only.columns)
```

```python
#separate input from output
inputRaw = data_NSP_only.drop('NSP', axis=1)
outputRaw = data_NSP_only['NSP']

display(inputRaw.head())
print(outputRaw[0:5])
```

```python
#plot correlation among inputs
pd.plotting.scatter_matrix(inputRaw, alpha = 0.3, figsize = (16,16), diagonal = 'kde');
```

```python
#plot out suspected highly correlated attributes on larger graph
suspectedCols = ['LBE','LB','Width','Min','Mode','Mean','Median']
pd.plotting.scatter_matrix(inputRaw[suspectedCols], alpha = 0.3, figsize = (16,16), diagonal = 'kde');
```

##### According to the above graph, LB and LBE are highly correlated with each other; min and width, and the group of mode, median, and mean also shows strong correlation. As a result, only one attribute will be seleted from each group in the purpose of reducing dimensionality. LB is selected based on the description, SisPorto, so that the selected attributes (AC, FM, UC, ASTV, mSTV, ALTV, mLTV) uniformally from SisPorto. Wdith and Median are selected because they are more uniformally distributed. #####

```python
dependents = ['LBE','Min','Mode','Mean']
inputIndependent = inputRaw.drop(dependents, axis=1)
    
display(inputIndependent.head())
print(inputIndependent.shape)
```

```python
#Plot scatter matrix again and observe which attributes need log transformation
pd.plotting.scatter_matrix(inputIndependent, alpha = 0.3, figsize = (16,16), diagonal = 'kde');
```

```python
#Columns need log transformation
logTranformCol = ['AC','FM','UC','MSTV','ALTV','MLTV','DL','Nzeros','Variance']
logData = inputIndependent[logTranformCol].apply(lambda x: np.log(x + 0.01))
inputWithLog = inputIndependent.drop(logTranformCol, axis=1)
inputWithLog = pd.concat([inputWithLog, logData], axis=1, join_axes=[inputWithLog.index])
display(inputWithLog.head())
```

```python
#Min-Max Scale input data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(inputWithLog)
scaledInput = scaler.transform(inputWithLog)

display(scaledInput)

#Save scaler model
import pickle
filename = 'scaler.sav'
pickle.dump(scaler, open(filename, 'wb'))
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaledInput, outputRaw, test_size=0.2, random_state=100)
y_train = y_train.as_matrix()
y_test = y_test.as_matrix()
```

```python
# Accuracy, Recall, and Precision based off Category 3
from sklearn.metrics import fbeta_score
def evaluation(y, pred, target=3, name='Unnamed', beta=2):
    result = {}
    
    acc = 0
    TP = 0
    FP = 0
    FN = 0
    
    for i in range(0,y.shape[0]):
        if y[i] == pred[i]:
            acc = acc + 1
            if y[i] == target:
                TP = TP + 1
        else:
            if y[i] == target:
                FN = FN + 1
            elif pred[i] == target:
                FP = FP + 1
    
    result['Name'] = name
    result['Accuracy'] = 1.0*acc/y.shape[0]
    result['TP'] = TP
    result['FP'] = FP
    result['FN'] = FN
    result['Count'] = y.shape[0]
    result['Recall'] = 1.0 * TP / (TP+FN)
    result['Precesion'] = 1.0 * TP / (TP+FP)
    result['FBeta'] = (1.0+beta*beta)*result['Precesion']*result['Recall']/(beta*beta*result['Precesion']+result['Recall'])
    result['FBetaAll'] = fbeta_score(y, pred, average='weighted', beta=beta)
    
    return result
```

```python
def custom_fbeta_score(y, pred, target=3, beta=2):
    acc = 0
    TP = 0
    FP = 0
    FN = 0
    
    for i in range(0,y.shape[0]):
        if y[i] == pred[i]:
            if y[i] == target:
                TP = TP + 1
        else:
            if y[i] == target:
                FN = FN + 1
            elif pred[i] == target:
                FP = FP + 1
                
    r = 1.0 * TP / (TP+FN)
    p = 1.0 * TP / (TP+FP)
    fb = (1.0+beta*beta)*p*r/(beta*beta*p+r)
    
    return fb
```

```python
# Benchmark Model
def benchmark(heartrate):
    if heartrate >= 120 and heartrate <= 160:
        return 1
    else:
        return 3
    
resultsBenchMark = {}

X_testBenchMark = scaler.inverse_transform(X_test)
y_predTestBenchMark = []
for i in range(0,X_testBenchMark.shape[0]):
    # heartrate median is at index 7
    y_predTestBenchMark.append(benchmark(X_testBenchMark[i][7]))

resultsBenchMark = evaluation(y_test, y_predTestBenchMark, name='BenchMark')

print(resultsBenchMark)
```

```python
# Classification
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

clfs = []
results = []
clfs.append(GaussianNB())
clfs.append(DecisionTreeClassifier(random_state=1))
clfs.append(LogisticRegression(random_state=1,multi_class="multinomial",solver="newton-cg"))
clfs.append(KNeighborsClassifier(n_neighbors=5))
clfs.append(RandomForestClassifier(max_depth=5, random_state=1))

for clf in clfs:
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    results.append(evaluation(y_test, y_pred,name=clf.__class__.__name__))
    
print(results)
```

```python
# Random Forest Fine Tuning
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import make_scorer

clf = RandomForestClassifier(random_state=1)
parameters = {'max_features':['sqrt', 'log2'],'max_depth':list(range(3,50)),'min_samples_split':[2,3,4,5]}

# Make an fbeta_score scoring object using make_scorer()
scorer = make_scorer(custom_fbeta_score, beta=3)

# # Perform grid search on the classifier using 'scorer' as the scoring method using GridSearchCV()
grid_obj = GridSearchCV(clf, parameters, scoring=scorer)

# # Fit the grid search object to the training data and find the optimal parameters using fit()
grid_fit = grid_obj.fit(X_train, y_train)

# # Get the estimator
best_clf = grid_fit.best_estimator_

# # Make predictions using the unoptimized and model
best_predictions = best_clf.predict(X_test)
print(best_clf.get_params())

best_result = evaluation(y_test, best_predictions)
print(best_result)
```

```python
# Create Uniform Data for Neural Network
X = X_train.tolist()
y = y_train.tolist()

for i in range(0,X_train.shape[0]):
    if y_train[i] == 2:
        for j in range(0,5):
            X.append(X_train[i])
            y.append(y_train[i])
    elif y_train[i] == 3:
        for j in range(0,9):
            X.append(X_train[i])
            y.append(y_train[i])
            
for i in range(0, len(y)):
    X[i] = np.append(X[i],y[i])

np.random.shuffle(X)

for i in range(0, len(y)):
    y[i] = X[i][18]
    
X_train_uniform = np.delete(X, 18, 1)
y_train_uniform = y
```

```python
# Transform output
def backtransform(r):
    y = []
    for i in range(0, r.shape[0]):
        if r[i][2] >= r[i][1] and r[i][2] >= r[i][0]:
            y.append(3)
        elif r[i][1] > r[i][2] and r[i][1] >= r[i][0]:
            y.append(2)
        else:
            y.append(1)
    return np.asarray(y)
```

```python
from keras import backend as K

def fbetaCat3(y_true, y_pred, threshold_shift=0, beta = 3):
    # just in case of hipster activation at the final layer
    y_pred = K.clip(y_pred, 0, 1)

    # shifting the prediction threshold from .5 if needed
    y_pred_bin = K.round(y_pred + threshold_shift)

    tp = K.sum(K.round(y_true * y_pred_bin), axis=1)
    fp = K.sum(K.round(K.clip(y_pred_bin - y_true, 0, 1)))
    fn = K.sum(K.round(K.clip(y_true - y_pred, 0, 1)))

    precision = tp / (tp + fp)
    recall = tp / (tp + fn)

    beta_squared = beta ** 2
    return (beta_squared + 1) * (precision * recall) / (beta_squared * precision + recall)
```

```python
# Neural Network
from keras.models import Sequential
from keras.layers.core import Dense, Activation

model = Sequential()
model.add(Dense(18, input_dim=X_train_uniform.shape[1]))
model.add(Activation('tanh'))
model.add(Dense(3))
model.add(Activation('sigmoid'))

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics = ['categorical_accuracy'])
model.summary()

outputWithDummy = pd.get_dummies(y_train_uniform).as_matrix()
```

```python
model.fit(X_train_uniform, outputWithDummy, epochs=1200, verbose=0, batch_size = 200)

score = model.evaluate(X_train_uniform, outputWithDummy)
print(score)
```

```python
nn_pred_proba_train = model.predict_proba(X_train)
nn_pred_train = backtransform(nn_pred_proba_train)
nn_result_train = evaluation(y_train, nn_pred_train, name = 'NeuralNetwork', beta = 3)

print(nn_result_train)

nn_pred_proba = model.predict_proba(X_test)
nn_pred = backtransform(nn_pred_proba)
nn_result = evaluation(y_test, nn_pred, name = 'NeuralNetwork', beta = 3)

print(nn_result)
```
